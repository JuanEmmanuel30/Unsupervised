```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Gera o dataset
df = pd.read_csv('Country-data.csv')

# Remove a coluna 'Country'
df_numero = df.drop(columns=['Country'])

# Padronizacao dos  dados
scaler = StandardScaler()
df_padronizado = scaler.fit_transform(df_numero)

# Aplicar PCA para reduzir a 2 dimensões
pca = PCA(n_components=2)
pca_resultado = pca.fit_transform(df_padronizado)

# Criacao do DataFrame
df_pca = pd.DataFrame(pca_resultado, columns=['PC1', 'PC2'])
df_pca['Country'] = df['Country']

# Visualizar com gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.7)
for i, pais in enumerate(df_pca['Country']):
    plt.text(df_pca['PC1'][i], df_pca['PC2'][i], pais, frontsize=6, alpha=0.6)
plt.title('Visualização dos países com PCA (2 componentes principais)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 1. Carregar o dataset
df = pd.read_csv('Country-data.csv')

# 2. Remover a coluna 'country' para trabalhar apenas com dados numéricos
df_numerico = df.drop(columns=['Country'])

# 3. Padronizar os dados para que todas as variáveis tenham a mesma escala
scaler = StandardScaler()
df_padronizado = scaler.fit_transform(df_numero)

# 4. Aplicar o algoritmo K-Médias com 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(df_padronizado)

# 5. Adicionar os rótulos dos clusters
df['cluster'] = clusters

# 6. Reduzir os dados para 2 dimensões com PCA
pca = PCA(n_components=2)
pca_resultado = pca.fit_transform(df_padronizado)

# 7.  Gráfico de dispersão com os clusters
plt.figure(figsize=(10, 6))
for grupo in range(3):
    plt.scatter(
        pca_resultado[df['cluster'] == grupo, 0],
        pca_resultado[df['cluster'] == grupo, 1],
        label=f'Grupo {grupo}'
    )
plt.title('Agrupamento dos países com K-Médias (3 grupos)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 8. Lista dos países agrupados
print("Países agrupados por cluster:")
print(df[['Country', 'cluster']].sort_values(by='cluster'))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# 1. Carregar o dataset
df = pd.read_csv('Country-data.csv')

# 2. Remover a coluna 'country' para trabalhar com dados numéricos
df_numerico = df.drop(columns=['Country'])

# 3. Padronizar os dados
scaler = StandardScaler()
df_padronizado = scaler.fit_transform(df_numero)

# 4. Aplicar clusterização hierárquica (método de ligação: 'ward')
ligacao = linkage(df_padronizado, method='ward')

# 5. Gerar dendrograma para visualizar os agrupamentos
plt.figure(figsize=(12, 6))
dendrogram(ligacao, labels=df['country'].values, leaf_rotation=90)
plt.title('Dendrograma - Clusterização Hierárquica dos Países')
plt.xlabel('País')
plt.ylabel('Distância')
plt.tight_layout()
plt.show()

# 6. Formar 3 grupos com base na hierarquia
grupos = fcluster(ligacao, t=3, criterion='maxclust')
df['cluster'] = grupos

# 7. Exibir os países agrupados
print("Países agrupados por cluster:")
print(df[['country', 'cluster']].sort_values(by='cluster'))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregar o dataset
df = pd.read_csv('Country-data.csv')

# 2. Separar a coluna de países
paises = df['Country']

# 3. Remover a coluna 'country' e padronizar os dados
df_numerico = df.drop(columns=['Country'])
scaler = StandardScaler()
df_padronizado = scaler.fit_transform(df_numero)

# 4. Aplicar K-Médias com 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df_padronizado)

# 5. Calcular a média das variáveis por cluster
medias_clusters = df.groupby('cluster').mean(numeric_only=True)

# 6. Visualizar a distribuição das dimensões por cluster
plt.figure(figsize=(12, 6))
sns.heatmap(medias_clusters.T, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Distribuição média das variáveis por cluster (K-Médias)')
plt.xlabel('Cluster')
plt.ylabel('Variável')
plt.tight_layout()
plt.show()

# 7. Mostrar a tabela de médias
print("Distribuição média das variáveis por cluster:")
print(medias_clusters)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

# 1. Carregar o dataset
df = pd.read_csv('Country-data.csv')

# 2. Separar a coluna de países
paises = df['Country']

# 3. Remover a coluna 'country' e padronizar os dados
df_numerico = df.drop(columns=['Country'])
scaler = StandardScaler()
df_padronizado = scaler.fit_transform(df_numero)

# 4. Aplicar K-Médias com 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df_padronizado)

# 5. Identificar o país mais próximo do centróide de cada cluster
representantes = {}
for i in range(3):
    indices = df[df['cluster'] == i].index
    cluster_dados = df_padronizado[indices]
    centroide = kmeans.cluster_centers_[i].reshape(1, -1)
    mais_proximo, _ = pairwise_distances_argmin_min(centroide, cluster_dados)
    pais_representativo = df.loc[indices[mais_proximo[0]], 'Country']
    representantes[i] = pais_representativo

# 6. Mostrar os países representativos por cluster
print("País que melhor representa cada cluster:")
for cluster_id, pais in representantes.items():
    print(f"Cluster {cluster_id}: {pais}")
```
